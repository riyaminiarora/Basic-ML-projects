```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.linear_model import LinearRegression
from  sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('Downloads/Graduate Admission.csv')
```

```python
ds=dataset.copy()
```

```python
ds.head()
```

```python
print(ds.shape)
print(ds.isnull().sum())
print(ds.info())
print(ds.describe())
```

```python
sns.pairplot(ds)
```

```python
correlation=ds.corr()
correlation
```

```python
ds2=ds.drop('Serial No.',axis=1)
ds2
```

```python
plt.boxplot(ds2["GRE Score"],showmeans=True)
plt.show()
plt.boxplot(ds2["TOEFL Score"],showmeans=True)
plt.show()
plt.boxplot(ds2["University Rating"],showmeans=True)
plt.show()
plt.boxplot(ds2["SOP"],showmeans=True)
plt.show()
#plt.boxplot(ds2["LOR"],showmeans=True)
#plt.show()
plt.boxplot(ds2["CGPA"],showmeans=True)
plt.show()
```

```python
x=ds2.drop(['Admit','SOP','University Rating'],axis=1)
y=ds2.Admit
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15,shuffle=True)
clf=LinearRegression()#classifier
clf.fit(x_train,y_train)#training
```

```python
clf.score(x_test,y_test)#check accuracy
```
